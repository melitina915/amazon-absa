# . , 처리


```python
# df['review'] = df['review_title'] + df['review_content']
# df.drop(columns = ['review_title', 'review_content'], inplace=True)

df.drop(columns = ['review_title'], inplace=True)

df
```

```python
# 1. review 칼럼을 쉼표(,) 기준으로 분리(split)
df['review_content'] = df['review_content'].str.split(',')

# 2. review 칼럼을 explode해서 각 문장이 한 행이 되도록 만듦
df = df.explode('review_content').reset_index(drop=True)

# 결과 확인
df
```

```python
# 1. review 칼럼을 . 기준으로 분리(split)
df['review_content'] = df['review_content'].str.split('.')

# 2. review 칼럼을 explode해서 각 문장이 한 행이 되도록 만듦
df = df.explode('review_content').reset_index(drop=True)

# 결과 확인
df
```

```python
# 1. review 칼럼을 . 기준으로 분리(split)
df['review_content'] = df['review_content'].str.split('!')

# 2. review 칼럼을 explode해서 각 문장이 한 행이 되도록 만듦
df = df.explode('review_content').reset_index(drop=True)

# 결과 확인
df
```

```python
# 1. review 칼럼을 . 기준으로 분리(split)
df['review_content'] = df['review_content'].str.split('?')

# 2. review 칼럼을 explode해서 각 문장이 한 행이 되도록 만듦
df = df.explode('review_content').reset_index(drop=True)

# 결과 확인
df
```

```python
import pandas as pd

# 특수문자만 포함된 행만 찾기 (한글, 영문, 숫자 제외)
mask = ~df['review_content'].str.match(r'^[^a-zA-Z0-9가-힣]+$')

df = df[mask].reset_index(drop=True)
df
```

```python
df.head(50)
```

```python
import pandas as pd

# 1. review가 결측치(NaN)이거나 공백만 있는 경우 삭제
df = df[df['review_content'].notnull()]                # NaN 제거
df = df[df['review_content'].str.strip() != '']        # 공백만 있는 행 제거

# 2. 한 단어만 있는 경우 삭제
# 단어 수가 2개 이상인 행만 남김
df = df[df['review_content'].str.strip().str.count(r'\S+') > 1].reset_index(drop=True)

df
```

# 가격 할인율

```python
# 가격 및 할인율 컬럼 전처리 함수
def clean_price_column(series):
    return series.replace('[₹,%]', '', regex=True).replace(',', '', regex=True).astype(float)

# 해당 컬럼들 변환
df['discounted_price'] = clean_price_column(df['discounted_price'])
df['actual_price'] = clean_price_column(df['actual_price'])
df['discount_percentage'] = clean_price_column(df['discount_percentage'])

# 변환 결과 확인
df[['discounted_price', 'actual_price', 'discount_percentage']].head()
```

```python
# '|'로 분리하여 첫 번째 요소만 추출
df['large_category'] = df['category'].apply(lambda x: x.split('|')[0])

# 결과 확인
df[['category', 'large_category']].head()
```
